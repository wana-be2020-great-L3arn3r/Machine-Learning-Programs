```python

import numpy as np
import matplotlib.pyplot as plt

# Real-life example: Predicting house prices based on size (in square feet)
house_sizes = np.array([600, 800, 1000, 1200, 1500, 1800, 2000, 2500])
house_prices = np.array([150000, 200000, 250000, 280000, 350000, 400000, 450000, 500000])

# Mean of x and y
mean_x = np.mean(house_sizes)
mean_y = np.mean(house_prices)

# Calculating slope (m) and intercept (b)
numerator = sum((house_sizes - mean_x) * (house_prices - mean_y))
denominator = sum((house_sizes - mean_x) ** 2)
m = numerator / denominator
b = mean_y - (m * mean_x)

# Predict prices using the manually computed regression model
predicted_prices = m * house_sizes + b

# Plot the results
plt.scatter(house_sizes, house_prices, color='red', label='Actual Prices')
plt.plot(house_sizes, predicted_prices, color='blue', linewidth=2, label='Predicted Prices')
plt.xlabel("House Size (sq ft)")
plt.ylabel("House Price ($)")
plt.legend()
plt.title("Linear Regression: House Size vs Price (Without sklearn)")
plt.show()

# Print manually calculated model parameters
print(f"Intercept: {b}")
print(f"Coefficient: {m}")
```
