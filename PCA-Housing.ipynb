## PCA Demonstration in Python

```python
# Importing the required libraries
```

```python
import numpy as np, pandas as pd
```

```python
data = pd.read_csv("newhousing.csv")
```

```python
data
```

```python
data.describe()
```

Scaling the data

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X = data.copy()
```

```python
x = scaler.fit_transform(X)
```

```python
dfx = pd.DataFrame(x, columns=X.columns)
dfx
```

### Applying PCA on the data

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(random_state=100)
```

```python
pca.fit(x)
```

```python
pca.components_
```

```python
pca.explained_variance_ratio_
```

```python
import matplotlib.pyplot as plt
```

```python
plt.bar(range(1,len(pca.explained_variance_ratio_)+1), pca.explained_variance_ratio_)
```

```python
var_cumu = np.cumsum(pca.explained_variance_ratio_)
```

#### Making the scree plot

```python
plt.plot(range(1,len(var_cumu)+1), var_cumu)
```

PCA with 2 components

```python
pc2 = PCA(n_components=6, random_state=100)
```

```python
newdata = pc2.fit_transform(x)
```

```python
newdata.shape
```

Making a dataframe out of it for convenience



```python
df = pd.DataFrame(newdata, columns=["PC1", "PC2", "PC3", "PC4", "PC5", "PC6"])
df.head()
```
