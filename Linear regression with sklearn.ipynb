```python
import numpy as np
import matplotlib.pyplot as plt
!pip install scikit-learn
from sklearn.linear_model import LinearRegression

# Real-life example: Predicting house prices based on size (in square feet)
house_sizes = np.array([600, 800, 1000, 1200, 1500, 1800, 2000, 2500]).reshape(-1, 1)
house_prices = np.array([150000, 200000, 250000, 280000, 350000, 400000, 450000, 500000])

# Create and train the model
model = LinearRegression()
model.fit(house_sizes, house_prices)

# Predict prices for given house sizes
predicted_prices = model.predict(house_sizes)

# Plot the results
plt.scatter(house_sizes, house_prices, color='red', label='Actual Prices')
plt.plot(house_sizes, predicted_prices, color='blue', linewidth=2, label='Predicted Prices')
plt.xlabel("House Size (sq ft)")
plt.ylabel("House Price ($)")
plt.legend()
plt.title("Linear Regression: House Size vs Price")
plt.show()

# Print model parameters
print(f"Intercept: {model.intercept_}")
print(f"Coefficient: {model.coef_[0]}")
```
